```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score
```

```python
# 데이터 읽기
df = pd.read_csv('network.csv')
```

```python
df.head()
```

```python
# 문자열 열을 연결
df['text_column'] = df['강점1'] + ' ' + df['강점2'] + ' ' + df['강점3'] + ' ' + df['강점4'] + ' ' + df['강점5']
```

```python
df.head()
```

```python
# 각 문장을 단어 집합으로 변환
sentences = df['text_column'].apply(lambda x: set(x.split()))

# 각 문장 간의 Jaccard 유사도를 계산
similarity = []
for i in range(len(sentences)):
    row = []
    for j in range(len(sentences)):
        row.append(jaccard_score(list(sentences[i]), list(sentences[j]), average='macro'))
    similarity.append(row)
```

```python
# 유사도 네트워크 생성
G = nx.from_numpy_array(np.array(similarity))
```

```python
#이제 네트워크 분석을 수행할 수 있습니다.
# 그래프에서 연결 중심성 계산
centrality = nx.degree_centrality(G)
```

```python
# 그래프 시각화
nx.draw(G, with_labels=True)
plt.show()
```

```python
# 그래프 시각화
#nx.draw(G, with_labels=True)
#plt.show()
plt.figure(figsize=(10, 10))  # 그래프 크기 조정
pos = nx.spring_layout(G)  # 노드의 위치를 Fruchterman-Reingold force-directed algorithm을 사용하여 배치

nx.draw_networkx_nodes(G, pos, node_size=500, node_color='blue')  # 노드 그리기
nx.draw_networkx_edges(G, pos, edge_color='grey')  # 엣지 그리기
nx.draw_networkx_labels(G, pos, font_size=12, font_color='white')  # 레이블 그리기

plt.axis('off')  # 축 끄기
plt.show()  # 그래프 보여주기
```
