```python
import numpy as np
from scipy import optimize
from scipy.optimize import Bounds

import matplotlib.pyplot as plt
import pandas as pd
```

```python
def calculate_total_distance(intercepts, x_data, y_data, slopes):
    total_distance = 0
    for x, y in zip(x_data, y_data):
        for i, m in enumerate(slopes):
            c = intercepts[i]
            # 직선 방정식: y = mx + c
            distance = abs(y - m*x - c) / np.sqrt(1 + m**2)
            total_distance += distance
    return total_distance

def calculate_initial_intercepts(points, slopes):
    initial_intercepts = []
    for point, slope in zip(points, slopes):
        x, y = point
        intercept = y - slope * x
        initial_intercepts.append(intercept)
    return initial_intercepts

def plot_lines_and_points(x_data, y_data, slopes, intercepts):
    x = np.linspace(min(x_data)-1, max(x_data)+1, 400)
    plt.figure(figsize=(8, 6))
    
    for m, c in zip(slopes, intercepts):
        y = m * x + c
        plt.plot(x, y)
        
    plt.plot(x_data, y_data, 'ro') # points를 red circle로 표시
    
    plt.xlim(min(x_data), max(x_data))
    plt.ylim(min(y_data), max(y_data))
    plt.grid(True)
    plt.show()

```

```python
#params
H0 = 0.0002
Lv = 0.001
LD = 0.001
Lf = 0.0005
```

```python
# line params
slopes = [np.tan(np.radians(130)),np.tan(np.radians(120)),np.tan(np.radians(95)),np.tan(np.radians(0))]
```

```python
data_hc=pd.read_csv('HC_sheet.csv',skiprows=[1])

left_idx = data_hc[data_hc['DOE0'] < 0.0005].index
x_data = data_hc['DOE0'].loc[left_idx]
y_data = data_hc['Unnamed: 1'].loc[left_idx].values*1E-6
```

```python
initial_points = [(-1*Lv+H0/np.tan(np.radians(50)), 0), (0, 0), (0.001, 0.0005), (0, 0)]  # 초기 절편을 결정하는 (x, y) 점들
initial_intercepts = calculate_initial_intercepts(initial_points, slopes)

# 각 절편에 대한 최소, 최대 값 설정
bounds = Bounds([-1*Lv+H0/np.tan(np.radians(50)), -0.001, 0, 0], [H0/np.tan(np.radians(50)), 0, 0.001, 0.001])

res = minimize(calculate_total_distance, initial_intercepts, args=(x_data, y_data, slopes), bounds=bounds, method='L-BFGS-B')
optimal_intercepts = res.x
plot_lines_and_points(x_data, y_data, slopes, optimal_intercepts)
```
